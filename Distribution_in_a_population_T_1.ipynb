```python
female_df = df[df["Series Name"] == "Population, female (% of total population)"]
male_df = df[df["Series Name"] == "Population, male (% of total population)"]
```

```python
female_df = df[df["Series Name"] == "Population, female (% of total population)"]
male_df   = df[df["Series Name"] == "Population, male (% of total population)"]

years = [str(y) for y in range(2001, 2023)]

female_top = female_df[female_df["Country Name"].isin(top_10_population_2022['Country Name'])]
male_top   = male_df[male_df["Country Name"].isin(top_10_population_2022['Country Name'])]
```

```python
for country in top_10_population_2022['Country Name']:
    f_row = female_top[female_top["Country Name"] == country][years].values.flatten()
    m_row = male_top[male_top["Country Name"] == country][years].values.flatten()

    plt.figure(figsize=(10, 5))
    plt.plot(years, f_row, label="Female %", linewidth=2)
    plt.plot(years, m_row, label="Male %", linewidth=2)
    plt.title(f"Male% vs Female% Across Years (2001–2022)\n{country}", fontsize=14)
    plt.xlabel("Year")
    plt.ylabel("Percentage of Total Population")
    plt.xticks(rotation=45)
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.show()
```

```python
female_df = df[df["Series Name"] == "Population, female"][["Country Name", "2022"]]
female_df = female_df.rename(columns={"2022": "Female"})

male_df = df[df["Series Name"] == "Population, male"][["Country Name", "2022"]]
male_df = male_df.rename(columns={"2022": "Male"})

# Filter only top 10
female_top10 = female_df[female_df["Country Name"].isin(top_10_countries)]
male_top10   = male_df[male_df["Country Name"].isin(top_10_countries)]

# Merge data
merged = pd.merge(female_top10, male_top10, on="Country Name").set_index("Country Name")
```

```python
plt.figure(figsize=(18, 18))

for i, country in enumerate(top_10_countries, 1):
    female = merged.loc[country, "Female"]
    male = merged.loc[country, "Male"]
    sizes = [female, male]

    plt.subplot(5, 2, i)   # 5 rows × 2 columns = 10 plots
    plt.pie(
        sizes,
        labels=["Female", "Male"],
        autopct="%1.1f%%",
        startangle=90
    )
    plt.title(f"Gender Share in {country} (2022)")

plt.tight_layout()
plt.show()
```

```python
female_df = df[df["Series Name"] == "Population, female"][["Country Name", "Country Code", "2022"]]
female_df = female_df.rename(columns={"2022": "Female"})

male_df = df[df["Series Name"] == "Population, male"][["Country Name", "Country Code", "2022"]]
male_df = male_df.rename(columns={"2022": "Male"})
```

```python
merged = pd.merge(female_df, male_df, on=["Country Name", "Country Code"])
merged = merged.dropna()
```

```python
merged["F_M_Ratio"] = merged["Female"] / merged["Male"]
```

```python
import plotly.express as px

fig = px.choropleth(
    merged,
    locations="Country Code",
    color="F_M_Ratio",
    hover_name="Country Name",
    color_continuous_scale="RdBu",
    title="World Map: Female-to-Male Population Ratio (2022)",
)

fig.update_layout(
    coloraxis_colorbar=dict(
        title="F/M Ratio",
        tickvals=[0.9, 1.0, 1.1, 1.2],
        ticktext=["0.9", "1.0", "1.1", "1.2"]
    )
)

fig.show()
```
